## Monthly Trend by Issue Date for Total Funded Amount

```python
# Ensure issue_date is datetime
data['issue_date'] = pd.to_datetime(data['issue_date'])

# Create monthly funded dataframe
monthly_funded = (
    data.sort_values('issue_date')
        .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %y'))
        .groupby('month_name', sort=False)['loan_amount']
        .sum()
        .div(1_000_000)
        .reset_index(name='loan_amount_millions')
)

# Create numeric x-axis for plotting
x = range(len(monthly_funded))

plt.figure(figsize=(10, 5))

# Area + line plot
plt.fill_between(
    x,
    monthly_funded['loan_amount_millions'],
    alpha=0.5
)
plt.plot(
    x,
    monthly_funded['loan_amount_millions'],
    linewidth=2
)

# Add data labels
for i, row in monthly_funded.iterrows():
    plt.text(
        i,
        row['loan_amount_millions'] + 0.1,
        f"{row['loan_amount_millions']:.2f}",
        ha='center',
        va='bottom',
        fontsize=9
    )

# Titles and labels
plt.title('Total Funded Amount by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Funded Amount (Millions)')

# X-axis labels
plt.xticks(x, monthly_funded['month_name'], rotation=45)

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
