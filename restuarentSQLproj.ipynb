### the third objective is to Analyze customer behavour

```python
## First thing to do is to combine  the tables
```

```python
q("""
SELECT 
    od.order_id,
    od.order_date,
    mi.item_name,
    mi.category,
    mi.price
FROM order_details od
LEFT JOIN menu_items mi
    ON od.item_id = mi.menu_item_id
""")
```

```python
q("""
SELECT 
    mi.item_name,
    mi.category,
    COUNT(od.order_id) AS total_orders,
    AVG(mi.price) AS avg_price
FROM order_details od
LEFT JOIN menu_items mi
    ON od.item_id = mi.menu_item_id
GROUP BY mi.item_name, mi.category
""")
```

```python
q("""
SELECT 
    mi.item_name,
    mi.category,
    COUNT(od.order_id) AS total_orders,
    AVG(mi.price) AS avg_price
FROM order_details od
LEFT JOIN menu_items mi
    ON od.item_id = mi.menu_item_id
GROUP BY mi.item_name, mi.category
ORDER BY total_orders DESC
""")
```
